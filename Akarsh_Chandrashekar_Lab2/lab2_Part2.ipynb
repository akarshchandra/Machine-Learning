```python
Y=df['Class']
X=df.drop(['Class'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
X_train_normalized=preprocessing.normalize(X_train)
```

```python
X_train_normalized=pd.DataFrame(X_train_normalized,columns=X.columns)
```

```python
X_test_normalized=preprocessing.normalize(X_test)
X_test_normalized=pd.DataFrame(X_test_normalized,columns=X.columns)
```

```python
X_train_normalized.shape
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Activation 
```

```python
model=Sequential()
model.add(Dense(64, activation='relu', input_dim=61))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train_normalized, y_train, epochs=200, verbose=False)
scores=model.evaluate(X_train_normalized, y_train, verbose=False)
print("Training Accuracy:%.2f%%\n"%(scores[1]*100))
scores=model.evaluate(X_test_normalized, y_test, verbose=False)
print("Testing Accuracy:%.2f%%\n"%(scores[1]*100))
model.summary()
```
